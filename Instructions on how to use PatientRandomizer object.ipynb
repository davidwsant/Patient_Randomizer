## This is just giving examples of how to use the PatientRandomzier object

```python
import pandas as pd
```

```python
# read in your data in a dataframe
fake_epilepsy_data = pd.read_csv('Fake epilepsy data for pseudocode.csv')
fake_epilepsy_data.head()
```

```python
# Make sure the patient_randomizer.py file is in your folder when you run this line
from patient_randomizer import *
```

```python
# Read your dataframe into an object
randomized_object = PatientRandomizer(fake_epilepsy_data)
```

```python
# fit the data using the parmaters given in the previous cell (below I will show you how to change them)
randomized_object.fit()
```

```python
# If you want to just get the list of patients, you can grab that using ".selected_patients"
randomized_object.selected_patients
```

```python
# If you want to just return a dataframe, use the get_df function on the object
output_df = randomized_object.get_df()
output_df
```

## Changing parameters

By default, these are the titles for the columns containing this data:
 - First Medication: 'Medication 1'
 - Second Medication: 'Medication 2'
 - Patient ID: 'patient id'

By default, it also does not replace samples and it takes 5 samples from each possible combination.

If you would like to change any of these parameters, you can do so when you create the object by putting the name of the variable you want to change and what you want it to change to. 

```python
# to get the names of the objects, you can use the dir function to call a created object
dir(randomized_object)
```

```python
# Here is an example where I have it only pick one from each category and NOT replace samples.
# Of course, when you only pick 1 replacement doesn't matter, but this will show you how to do it. 
# I also put the names of the columns so you could see how to change these. 

random_object = PatientRandomizer(fake_epilepsy_data, number_samples=1, replacement=True, med1_title='Medication 1', med2_title='Medication 2', patient_title = 'patient id') 
random_object.fit()
output_df = random_object.get_df()
output_df
```
